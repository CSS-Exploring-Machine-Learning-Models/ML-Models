```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

df = pd.read_csv('SalaryData.csv', usecols=['Age', 'Years of Experience', 'Salary'])

x = df[["Years of Experience"]]
y = df["Salary"]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

model = LinearRegression()
model.fit(x_train, y_train)

plt.figure(figsize=(10, 6))
plt.scatter(x_train, y_train, color='blue', label='Training Data')
plt.scatter(x_test, y_test, color='green', label='Testing Data', marker='x')
plt.plot(x, model.predict(x), color='red', label='Regression Line')

equation = f'y = {model.intercept_:.2f} + {model.coef_[0]:.2f}x'
plt.text(0.95, 0.15, equation, fontsize=12, transform=plt.gca().transAxes, ha='right', va='bottom')

r_squared_train = model.score(x_train, y_train)
r_squared_test = model.score(x_test, y_test)
plt.text(0.95, 0.10, f'R^2 (Train): {r_squared_train:.6f}', fontsize=10, transform=plt.gca().transAxes, ha='right', va='bottom')
plt.text(0.95, 0.05, f'R^2 (Test): {r_squared_test:.6f}', fontsize=10, transform=plt.gca().transAxes, ha='right', va='bottom')

plt.title('Years of Experience vs. Salary')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend()
plt.grid(True)
plt.show()
```
